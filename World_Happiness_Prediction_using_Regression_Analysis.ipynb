### Single Variable Regression Model
Now that we have selected our prediction variables, we can generate a regression model that can predict happiness based off the our prediction variables. To best model the relationship between the predictor and the target, I  will include regression models up to the third order polynomial.

##### 2. Create a regression model.
To reduce redundant code, we will create a regression model function that takes in as arguments a dataframe of predictor variables, a target variable, a predictor variable and returns a statsmodels.ols regression model object.

```python
import statsmodels.formula.api as smf
def regress_model(df,target,predictor):
    poly_1_formula = target + ' ~ 1 + ' + predictor
    poly_2_formula = target + ' ~ 1 + ' + predictor + ' + I(' + predictor + ' ** 2.0)'
    poly_3_formula = target + ' ~ 1 + ' + predictor + ' + I(' + predictor + ' ** 2.0) + I('+ predictor + ' ** 3.0)'    
    
    # 1-st order polynomial
    poly_1 = smf.ols(formula=poly_1_formula, data=df.dropna()).fit()

    # 2-nd order polynomial
    poly_2 = smf.ols(formula=poly_2_formula, data=df.dropna()).fit()

    # 3-rd order polynomial
    poly_3 = smf.ols(formula=poly_3_formula, data=df.dropna()).fit()
    
    return poly_1, poly_2, poly_3
```

##### 3. Create subplots for every world prdictor variable and fit polynomial regression lines.
We would also like to visualize each regression model. We can do this by creating subplots for each predictor variable and fitting polynomial regression lines to it.

```python
def visualize_regression(df):
    
    fig = plt.figure(figsize=[20,20])
    fig.subplots_adjust(hspace=0.4, wspace=0.2)
    
    subplot_index = 1
    r_squared = 0
    for predictor in df.columns:
        if predictor != 'LL':            
            ax = plt.subplot('32'+str(subplot_index))
            subplot_index += 1

            ax.scatter(df[predictor], df['LL'], alpha=0.5)
            ax.set_xlabel(predictor)
            ax.set_ylabel('Life Ladder')

            poly_1, poly_2, poly_3 = regress_model(df,'LL',predictor)
            x = pd.DataFrame({predictor: np.linspace(df[predictor].min(), df[predictor].max(), len(df[predictor]))})

            ax.plot(x, poly_1.predict(x), 'y-', label='Poly n=1 $R^2$=%.2f' % poly_1.rsquared)
            ax.plot(x, poly_2.predict(x), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2.rsquared)
            ax.plot(x, poly_3.predict(x), 'r-', label='Poly n=3 $R^2$=%.2f' % poly_3.rsquared)           

            ax.legend(loc='best', prop={'size':'small'})
            
    
    plt.show()
```

### Multiple Regression Model
Now that we have looked at single variable regression, it is important that we combine all our predictor variables into one model and see they affect our target variable.

To achieve this the following steps must be performed in order:
1. Create a multiple regression formula
2. Create a multiple regression model
3. Interpret the data

```python
def create_regress_formula(target, df):
    formula = target + ' ~ 1'
    for predictor in df.columns:
        if predictor != target:
            formula = formula + ' + ' + predictor
    return formula
```

```python
def mulitple_regression(df, target):
    formu = create_regress_formula(target, df)
    model = smf.ols(formula=formu, data=df.dropna()).fit()
    return model
```

#### Residual Plot
To calculate the residuals we must first produce the model.

```python
def residual_plot(df, target, predictors, model):
    Y_pred = model.predict(df[predictors])
    residual = df[target].values-Y_pred
    plt.figure(figsize=[10,10])
    plt.scatter(df[target],residual, alpha=0.6)
    plt.xlabel(target)
    plt.ylabel("residual")
    plt.show()
```
